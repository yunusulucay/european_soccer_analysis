```python
# So we joined two tables. Make some analysis
joined_db = read_sql("select player_name, birthday, height, weight, \
overall_rating, potential from Player P \
join Player_Attributes PA on P.player_api_id=PA.player_api_id \
group by P.player_name", conn)
```

```python
jdb_count = read_sql("select count(*) from sqlite_schema \
where name='joined_table'", conn)
jdb_count = jdb_count.values[0][0]
```

```python
# I created a new db to work on it
if not jdb_count>=1:
    joined_db.to_sql("joined_table", conn, index=False)
```

```python
# Check for missing values as you see we dont have missing values

read_sql("select count(*) from additional_table \
where player_name is null", conn)
```

```python
# Check for missing for all columns

read_sql("select count(*)-count(player_name) missing_player_name, \
count(*)-count(birthday) missing_birthday, \
count(*)-count(height) missing_height, \
count(*)-count(weight) missing_weight, \
count(*)-count(overall_rating) missing_overall_rating, \
count(*)-count(potential) missing_potential \
from additional_table", conn)
```

We get the youngest and the oldest players

```python
read_sql("select min(age), max(age) from additional_table", conn)
```

```python
read_sql("select * from additional_table \
where age=22 or age=54", conn)
```

```python
read_sql("select * from additional_table \
where player_name like 'aa%' and age <30", conn)
```

```python
# We will compare age potential with height

read_sql("select round(avg(height),2) AVG_HEIGHT, \
round(avg(age)) AVG_AGE, \
overall_rating OVERALL_RATING from additional_table \
group by overall_rating \
order by overall_rating desc limit 5", conn)
```

```python
# Query minimum age and maximum age using subquery

read_sql("select * from additional_table \
where age=(select max(age) from additional_table) \
or age=(select min(age) from additional_table)", conn)
```
